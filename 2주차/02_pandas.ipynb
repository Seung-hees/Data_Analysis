```python
url = 'http://bit.ly/dsa-01-order'

order = pd.read_csv(url,index_col='id')

print(order.shape)
order.head()
```

# 집계하기

## Pivot table

```python
#state 를 나누어서 value 인 price 의 값들의 평균값을 나타내줌.(default 가 평균으로 설정되어 있음.)
#index 는 기준 / value 는 요약값
pd.pivot_table(data=order, index='state', values='price')
```

```python
pd.pivot_table(data=order, index='state', values='price', aggfunc=['mean','sum'])
```

```python
pd.pivot_table(data=order, index = ['date','state'], values ='price', aggfunc=['mean','sum'])
```

## groupby

```python
#같은 값을 가지고 있는 것들끼리 그룹을 만들어준다.

order.groupby('state').groups 
```

```python
order.groupby('state').get_group('canceled')
```

```python
group = order.groupby('state')

# group.groups
# group.get_group('canceled')
```

```python
group['price'].mean()
```

```python
group['price'].sum()
```

```python
#value 를 여러개 하고 싶을 때

group[['price','price']].mean()
```

```python
# index 를 여러개 사용하고 싶을 때

#order.groupby(['date','state'])['price'].mean()
group = order.groupby(['date','state'])
group['price'].mean()
```

# 데이터 병합하기

## concat

```python
url = 'http://bit.ly/dsa-02-order-201701'
order_01 = pd.read_csv(url)

print(order_01.shape)
order_01.head()
```

```python
url = 'http://bit.ly/dsa-02-order-201702'
order_02 = pd.read_csv(url)

print(order_02.shape)
order_02.head()
```

```python
#세로로 붙이기

order = pd.concat([order_01,order_02])

order = order.reset_index()
order
```

```python
#가로로 붙이기

pd.concat([order_01,order_02],axis=1)
```

## merge

```python
url = 'http://bit.ly/dsa-02-user'
user = pd.read_csv(url)

print(user.shape)
user.head()
```

```python
url = 'http://bit.ly/dsa-02-product'
product = pd.read_csv(url)

print(product.shape)
product.head()
```

```python
order.head(1)
```

```python
user.head(1)
```

```python
# inner (교집합) / outer (합집합) / left (왼쪽 기준) / right (오른쪽 기준)
```

```python
# 공통점이 있는 것만 있고, 없으면 삭제

pd.merge(order, user, left_on='user_id', right_on='id',how='inner')
```

```python
# 없어도 데이터를 불러옴.(order 데이터는 그대로 두고, user 데이터가 붙음)

pd.merge(order, user, left_on='user_id', right_on='id',how='left')
```

# apply

```python
def above_500(price):
    return price>500
```

```python
above_500(100)
```

```python
order['price'].apply(above_500)
```

```python
# vip - price > 500 & state == 'confirmed'

def is_vip(row):
    if row['price'] > 500 and row['state'] == 'confirmed' :
        return 'VIP'
    else: 
        return 'non-VIP'
```

```python
order.apply(is_vip, axis=1)
```
