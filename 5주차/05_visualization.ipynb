```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

from IPython.display import set_matplotlib_formats


# 그림을 jupyter notebook 안에 그리는 설정
%matplotlib inline

#판다스 max columns display 옵션
pd.options.display.max_columns = 100
```

```python
#한글 폰트 설정
mpl.rc('font' , family = 'NanumGothic')

#그림을 선명하게 설정
set_matplotlib_formats('retina')
```

```python
data = pd.read_csv('health_2016.csv')

print(data.shape)
data.head()
```

```python
sns.countplot (data=data, x= '성별코드')
```

# 그림 설정하기

## 그림 크기 조절하기

```python
# 범주형 데이터에 대해서 count 해주는 것이 좋음

plt.figure(figsize =(15,3))#(가로 길이, 세로길이)
sns.countplot(data = data,x='시도코드')
```

## 그림 여러개 한번에 그리기

```python
#fig - 전체 그림
#ax -subplot


# 2x2(행,렬)
fig, axes = plt.subplots(nrows= 1, ncols =2)
fig.set_size_inches(15, 3)

sns.countplot(data=data, x= '성별코드', ax=axes[0])
sns.countplot(data=data, x= '시도코드', ax=axes[1])
```

# distplot

```python
# distplot - 연속형 데이터의 분포
# 히스토그램의 y : 빈도 수
# 그래프의 y : 값

flg,axes =plt.subplots(nrows =1,ncols =3)
flg.set_size_inches(12,3)

sns.distplot(data['허리둘레'], ax=axes[0])
sns.distplot(data['허리둘레'], kde =False,ax=axes[1])
sns.distplot(data['허리둘레'], hist= False ,ax=axes[2])
```

```python
# hue 를 제공하고 있지 않음
# hue = '흡연상태'

# data['흡연상태'].unique()

w_1 = data.loc[data['흡연상태'] == 1,'허리둘레']
w_2 = data.loc[data['흡연상태'] == 2,'허리둘레']
w_3 = data.loc[data['흡연상태'] == 3,'허리둘레']

sns.distplot(w_1 , hist = False , label = '비흡연')
sns.distplot(w_2 , hist = False , label = '금연')
sns.distplot(w_3 , hist = False , label = '흡연')
```

# barplot

```python
# barplot = x(범주형)에 대한 y의 분포; 평균 & 신뢰구간

plt.figure(figsize = (10,4))
sns.barplot(data=data, x='연령대코드(5세단위)', y='수축기혈압',hue='성별코드')
```

```python
# estimator - 원하는 계산 방법
# ci - confidence interval(신뢰구간)


plt.figure(figsize = (10,4))
sns.barplot(data=data, x='연령대코드(5세단위)', y='수축기혈압', estimator = np.sum, ci = None)
```

```python
# horizontal 방향으로 그리기 orient = 'h'

plt.figure(figsize=(3,6))
sns.barplot(data=data, x='수축기혈압', y='연령대코드(5세단위)', orient='h')
```

# Pointplot

```python
# barplot 과 거의 동일한 의미
# 기울기 / 변화율 -> x가 연속 값을 가질 때 (ex . 시간 순)

data2 = data[data['연령대코드(5세단위)']<=17]

sns.pointplot(data=data2, x='연령대코드(5세단위)',y='수축기혈압',hue='성별코드')
```

# boxplot

```python
# x에 대한 y의 분포

plt.figure(figsize=(10,4))

sns.boxplot(data=data, x='연령대코드(5세단위)',y='수축기혈압',hue='성별코드')
```

## violinplot

```python
# x 각각에 대하여 y의 분포를 직접 그려주는 그림

plt.figure(figsize=(10,4))

sns.violinplot(data=data, x='연령대코드(5세단위)', y='수축기혈압', hue='성별코드')
```

```python
plt.figure(figsize=(10,4))

sns.violinplot(data=data, x='연령대코드(5세단위)', y='수축기혈압', hue='성별코드', split=False)
```

```python
# countplot , distplot = x의 분포 (범주형 / 연속형)
# (barplot / pointplot ) / (boxplot / violinplot) - x 각각에 대해 y의 분포 (x는 범주형, 단순 집계 / 분포)


flg,axes =plt.subplots(nrows =4)
flg.set_size_inches(15,8)

sns.barplot(data= data, x = '연령대코드(5세단위)',y='수축기혈압',hue='성별코드',ax=axes[0])
sns.pointplot(data= data, x = '연령대코드(5세단위)',y='수축기혈압',hue='성별코드',ax=axes[1])
sns.boxplot(data= data, x = '연령대코드(5세단위)',y='수축기혈압',hue='성별코드',ax=axes[2])
sns.violinplot(data= data, x = '연령대코드(5세단위)',y='수축기혈압',hue='성별코드',ax=axes[3], split=True)

```
